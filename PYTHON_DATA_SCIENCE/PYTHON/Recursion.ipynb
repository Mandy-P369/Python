```python
def show(num) : 
    if(num==0) : # it is base case
        return 
    print(num,end=" ")
    show(num-2.5)
    
num =int(input("Enter the number"))
show(num)
```

```python
def fact(n):
    if(n==0 or n==1):
        return 1
    else :
        return (n)*fact(n-1)
n=int(input("Enter the number"))
output=fact(n)
print(output)
```

```python
def bridge(list,idx=0) : 
    if(idx==len(list)):
        return
    print(list[idx])
    bridge(list,idx+1)
bridges = ["Beam","Girder","Slab","Truss","Arch","Cantilever","Suspension","Cable Slayed","Bowstring","Tied Arch"]
bridge(bridges)
```
