```python
# Create functions to get the average of three number.
def calc_average(a,b,c):
    return (a+b+c)/3
a = int(input("Enter the number a"))
b = int(input("Enter the number b"))
c = int(input("Enter the number c"))
output  = calc_average(a,b,c)
print(output)
```
