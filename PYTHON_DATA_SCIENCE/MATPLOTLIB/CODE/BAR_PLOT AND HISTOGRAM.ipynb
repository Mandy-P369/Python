```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import random
from io import StringIO,BytesIO
```

```python
num = np.arange(0,np.pi*4,step=1)
sin_x = np.sin(num)
# cos_x = np.cos(num)
plt.bar(num,sin_x,color='g',linestyle='--',edgecolor='black',linewidth=1)
# plt.bar(num,cos_x,color='r')
plt.title('Bar Plot')
plt.xlabel('num values')
plt.ylabel('sin_x values')
plt.show()
```
