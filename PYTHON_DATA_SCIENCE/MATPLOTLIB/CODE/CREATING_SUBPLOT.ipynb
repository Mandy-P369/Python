```python
import matplotlib.pyplot as  plt
%matplotlib inline
import pandas as pd
import numpy as np
import random
from io import StringIO,BytesIO
```

#### Creating the sub plot 

```python
x = [12,32,23,43,45,55,67,78,44]
y = [34,32,56,54,67,76,78,98,90]
plt.scatter(x,y,c='r')
plt.plot(x,y)
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Creating subplot : 
plt.subplot(2,2,1) 
plt.plot(x,y,'r',markersize=10)
plt.subplot(2,2,2)
plt.plot(x,y,'b',linestyle='dashed')
plt.show()
```
