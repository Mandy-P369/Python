```python
import matplotlib.pyplot as  plt
%matplotlib inline
import pandas as pd
import numpy as np
import random
from io import StringIO,BytesIO
```

#### Creating the sub plot 

```python
x = [12,32,23,43,45,55,67,78,44]
y = [34,32,56,54,67,76,78,98,90]
plt.scatter(x,y,c='r')
plt.plot(x,y)
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Creating subplot : 
plt.subplot(2,2,1) 
plt.plot(x,y,'r',markersize=10)
plt.subplot(2,2,2)
plt.plot(x,y,'b',linestyle='dashed')
plt.show()
```

```python
list1 = [90,12,89,23,78,34,67,45,67,56]
list2 = [12,23,34,45,56,67,78,89,90,97]
plt.scatter(list1,list2,c='g')
plt.plot(x,y,'r',linestyle='dashed')
plt.title('Example 3')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
plt.plot(x,y,2,2,1,markersize=10)
plt.show()
plt.plot(x,y,2,2,2,'b',linestyle='dashed')
plt.show()
```

```python
a = np.arange(0,np.pi*4,step=1)
sin_x = np.sin(a)
plt.scatter(a,sin_x,c='b')
plt.plot(a,sin_x,linestyle='dashed',markersize=20)
plt.title('Sin Graph')
plt.xlabel('Value a ')
plt.ylabel('Value sin_x')
plt.show()
```

```python
# Cosine graph :
b = np.arange(-1,np.pi*4,step=1)
cos_x = np.cos(b)
cos_x
plt.scatter(b,cos_x,c='r')
plt.title('Cosine Graph')
plt.xlabel('b Value')
plt.ylabel('cos_x value')
plt.plot(b,cos_x,linestyle='dotted',markersize=20)
plt.show()
```
