```python
# If the number of sheet is there inside the excel ,then we can retrieve the sinle sheet also ..

# @ Header : that contains the row names 
# @ names : names = ['Names','batch','Rollno','college'] :It will clarify the column name insie the table ..
# @ index_col : It will show the index mark column as first column ...
# @ uscols  : it will read only selected columns ..
# @ dtype : set data types 
# @ skiprows :  skip specific rows from the table ..
# @ nrows : read only first N rows ..
# @ na_values : if there is any value where there is nothing written over there then we will use na_values
#                to fill here like null,-

data = 'C:\Fetch_Excel.xlsx'
dd = pd.read_excel(data,sheet_name='Sheet1',header=0,names=['name','Stream','enrollmentno','College']
                  ,index_col=0,usecols=['name','Stream'],dtype={'Rollno':int})
dd = pd.read_excel(data,dtype={'Rollno' : int})
dd = pd.read_excel(data,nrows=1)
```
