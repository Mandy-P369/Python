```python
# 1 ....df.rename Function : data.rename(mapper=None,index=None,Columns=None,axis=None,copy=True,inplace=True,errors='ignore')
data.rename(columns={'first_name': 'name'}, inplace=True)
data.rename(columns={'last_name':'surname'},inplace =True)
data.rename(columns={'annualfinalpay':'finalpayment'},inplace = True)
data
```

```python
#2.... df.drop Function : It is used to remove the rows and columns .
# data.drop(labels=None,axis=0,index=None,columns=None,level=None,inplace=False,errors='raise')
data.drop(columns='discount',errors='raise')
```

```python
#3..... data.astype() : This function is used to change the data type.
# data[''].astype(dtype,copy=True,errors='raise')
data['netpay'].astype(int,errors='raise')
```

```python
#4.... data.apply() : It is used to apply the function to each rows and column...

# Applying function to each columns ..
data['salary'].apply(lambda x: x).sum()
data['netpay'].apply(lambda x:x*x).sum()
```

```python
# data.map() : works only on single columns not on whole data frame ...
data['bonus']= data['salary'].map(lambda x : x*2)
data
```

```python
# data.replace() : IT is used to replace the specific value inside the data frame....
# data.replace(to_replace = None,inplace = True,value = None, limit = None ,regex = True,method = None )
data.replace (to_replace = 'salary',value='netpay')
```

```python
# data.clip() : It is use to limit the value inside the data frame or series  within the maximum  and 
# minimum threshold ....
# data.clip(lower=None,upper=None,axis=None,inplace=False)
data.loc[101:200,['salary']].clip()

data
```
