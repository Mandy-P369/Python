```python
names = ["Aayush",'Aman',"disha",'Jeevni','Akshay','Sapna','Sonam','Vaishnavi','Muskan','Surendra','Veer','krishna','Rajendra','Rohit','Rahul','Oshi','Deeksha','Pushpenddra','jeevan','Ramprasad']
cities = ['Srinagar','Shimla','chandigarh','Lucknow','Leh','Bhopal','Patna','Raipur','Mumbai','Banglore','Gandhinagar','Chennai','Thiruvananthpuram','Bhubneshwar','Vishakhapatnam','Surat','Aizawl','Dispur','Itanagar','Shillog','Agartala','Imphal']
meta = []
for i in range(1,101):
    name = random.choice(names + [None])
    age = random.choice([random.randint(18,32),None])
    marks = random.choice([random.randint(40,100),None])
    city = random.choice(cities+["",None])
    meta.append([i,name,age,marks,city])
meta = pd.DataFrame(meta,columns = ['ID',"Name","Age","Marks","City"])
meta
```

```python
# fillna() : The fillna() function in Pandas is used to fill or replace missing values (NaN/None) in a DataFrame or Series with a specific value, method, or calculated value.
# It detects the missing values from our data ...
# value=number/Text : if we do then it will replace the None value with suggestive number/Text...
# inplace = true : Modify the original DataFrame or Series directly instead of returning a new one...
# method = 'ffill' forward fill / 'bfill' backward fill...
# axis = 0 (column wise) / 1 (row wise)...
# downcast = Try to reduce the data type size ....
# limit : It fill only the first N missing values ...
data.fillna(value=5,inplace = True)
meta.fillna(value=5)
```
