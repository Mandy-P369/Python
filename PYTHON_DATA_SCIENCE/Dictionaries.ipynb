```python
# Dictionaries : Dictionary is the collection of unordered , changeable  and indexed ,
# Dictionaries are written  with curly brackets , and they have key and values ..
```

```python
dict = {}
type(dict)
```

```python
dict = {12,21,23,34,34}
type(dict)
```

```python
dict = {"name":"aayush","designation":"data scientist","company":"Infosys"}
print(dict)
```

```python
print("The type of variable is : {}",type(dict))
```

```python
# We can only access value based on the key values 
print(dict["designation"])
print(dict["name"])
```

```python
# We can even loop through the dictionaries values ..
print(len(dict))
```

```python
for i in dict:
    print("The values in key",i,"is",dict[i])
```

```python
# We can also check both keys and values ..
for x in dict.items():
    print(x)
```

```python
# Adding new items inside the dictionaries 
dict['surname'] = 'pateriya'
```

```python
print(dict)
```

```python
for x in dict.items():
    print(x)
```

```python
# How to create the nested dictionary ...
details1 = {'location1':'sagar'}
details2 = {'location2':'Lalitpur'}
details3 = {'location3':'Jhansi'}
information = {'detail1':details1,'detail2':details2,'detail3':details3}
print(information)
```

```python
print(information['detail1']['location1'])
```

```python
print(information['detail2']['location2'])
```

```python
print(details1['location1'])
```

```python
Information2 = {'information1': information}
```
