```python
import pandas as pd

data = pd.DataFrame({
    "student": ["A","B","C","D","E","F","G","H","I","J"],
    "math": [78, 45, 89, 66, 92, 55, 71, 84, 90, 62],
    "science": [82, 50, 94, 60, 88, 58, 70, 80, 91, 65],
    "gender": ["M","F","M","F","M","F","M","F","M","F"]
})
data
```

```python
data.head()
```

```python
data.corr(numeric_only=True)
```

```python
sns.heatmap(data.corr(numeric_only=True))
```

```python
# jointplot jointplot allows to study the relationship between 2 numeric variables .
# the central chart display their correlation , it is usually a scatterplot , a hexbin plot,
# a 2D histogram , or 2D density plot .
# This plot is use to do the univariate analysis.......
```

```python
jointplot1 = sns.jointplot(x='math',y='science',data=data,kind='hex')
jointplot1
```

```python
jointplot2 = sns.jointplot(x='science',y='math',data=data,kind='hex')
jointplot2
```

```python
sns.jointplot(x='math',y='science',data=data,kind='reg')
```

```python
jointplot4 = sns.jointplot(x='science',y='math',data=data,kind='reg')
jointplot4 
```

# Pairplot :
### It is also known as scatter plot , in which one variable is in the same data row is matched with the 
### another variable's values , like this , pair plot are just elaborations on this, showing all variables paired with all other 
### variables .

```python
pairplot1 = sns.pairplot(data)
pairplot1
```

```python
pairplot2 = sns.pairplot(data,hue='gender')
```

```python
pairplot3 = sns.pairplot(data,hue='student')
```
