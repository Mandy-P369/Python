```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import random
from io import StringIO , BytesIO
```

```python
import pandas as pd

data = pd.DataFrame({
    "student": ["A","B","C","D","E","F","G","H","I","J"],
    "math": [78, 45, 89, 66, 92, 55, 71, 84, 90, 62],
    "science": [82, 50, 94, 60, 88, 58, 70, 80, 91, 65],
    "gender": ["M","F","M","F","M","F","M","F","M","F"]
})
data
```

```python
data.head()
```

```python
# Count plot : This plot is use to show the count of  the column ....
sns.countplot(data=data,x ='math',hue='science')
plt.show()
```

```python
sns.countplot(y='gender',data=data)
plt.show()
```

```python
data.loc[5,'gender'] = 'M'
```

```python
data.loc[1:4,'gender']='F'
```

```python
data
```

```python
sns.countplot(y='gender',data=data)
plt.show()
```

###  Bar plot :

```python
sns.barplot(data=data,y='student',x='gender')
plt.show()
```

```python
sns.barplot(data=data,y='gender',x='student')
plt.show()
```

### Box plot .....

```python
data
```

```python
sns.boxplot(data=data,x='gender',y='student')
plt.show()
```

```python
sns.boxplot(data=data,x='gender',y='math')
plt.show()
```

```python
sns.boxplot(data=data,x='gender',y='science',palette ='rainbow',orient='v')
plt.show()
```

```python
sns.boxplot(data=data,orient='v')
plt.show()
```

```python
sns.boxplot(y='gender',x='student',hue='science',data=data,orient='h')
plt.show()
```

## Violin plot : 
#### This plot helps us to see the distribution of data  in terms of kernel density estimation  anad the box plot ....


```python
sns.violinplot(data=data,x='gender',y='student',palette='rainbow')
plt.show()
```
