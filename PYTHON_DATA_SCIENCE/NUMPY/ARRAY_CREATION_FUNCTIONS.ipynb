## Array Creation Functions 

```python
 np.array()          : Create array from list or tuple np.array([1,2,3])
 np.arange()         : Create sequence with step np.arange(0,10,2)
 np.linspace()       : Evenly spaced values np.linspace(0,1,5)
 np.zeros()          : All zeros np.zeros((3,3))
 np.ones()           : All ones np.ones((2,2))
 np.eye()            : Identity matrix np.eye(3)
 np.full()           : Constant array np.full((2,2),7)
 np.random.rand()    :Random (0â€“1) np.random.rand(3,2)
 np.random.randint() : Random integers np.random.randint(1,10,(3,3))
 np.empty()          :Uninitialized array np.empty((3,3)
```

##### Copy function() and BroadCasting() :

```python
# np.copy() is used to create a complete copy of a NumPy array.
# It creates a new array in memory, so that changes in the new array do not affect the original array.
```
